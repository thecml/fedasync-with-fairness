# Imports

```python
import os.path

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import idx2numpy
import os
from pathlib import Path
```

# Paths

```python
curr_dir = os.getcwd()
root_dir = Path(curr_dir).absolute().parent.parent

_cifar10_base_path = Path.joinpath(root_dir, "data/partitioned_datasets/cifar10/lda")
cifar10_concentration_client_data = lambda x, y, z : str(_cifar10_base_path) + "\\concentration_{}\\Client-{}\\{}.npy".format(x, y, z)

_mnist_base_path = Path.joinpath(root_dir, "data/partitioned_datasets/mnist/lda/")
mnist_concentration_client_data = lambda x, y, z : str(_mnist_base_path) + "\\concentration_{}\\Client-{}\\{}.npy".format(x, y, z)
```

# Utility

### Dataloading and formatting

```python
def load_mnist_client_labels(clientId, distribution):
    train_labels = np.load(mnist_concentration_client_data(distribution, clientId, "train_labels"))
    val_labels = np.load(mnist_concentration_client_data(distribution, clientId, "val_labels"))
    test_labels = np.load(mnist_concentration_client_data(distribution, clientId, "test_labels"))
    return np.concatenate([train_labels, val_labels, test_labels])
    
def load_cifar10_client_labels(clientId, distribution):
    train_labels = np.load(cifar10_concentration_client_data(distribution, clientId, "train_labels"))
    val_labels = np.load(cifar10_concentration_client_data(distribution, clientId, "val_labels"))
    test_labels = np.load(cifar10_concentration_client_data(distribution, clientId, "test_labels"))
    return np.concatenate([train_labels, val_labels, test_labels])
```

# Visualize

### MNIST post-LDA label distributions

##### alpha=1

```python
alpha=1

class_data = []
for client in range(1, 11):
    client_labels = load_mnist_client_labels(client, alpha)
    amount_of_class = []
    for label in range(10):
        amount_of_class.append(np.count_nonzero(client_labels==label))
    class_data.append(amount_of_class)
class_data = np.swapaxes(class_data, 0, 1)
clients = ['Client 1', 'Client 2', 'Client 3', 'Client 4', 'Client 5', 'Client 6', 'Client 7', 'Client 8', 'Client 9', 'Client 10']
classes = ['Class 0', 'Class 1', 'Class 2', 'Class 3', 'Class 4', 'Class 5', 'Class 6', 'Class 7', 'Class 8', 'Class 9']

fig, ax = plt.subplots()
ax.invert_yaxis()

x_sum = np.zeros(10)
for i in range(10):
    ax.barh(
            clients,
            class_data[i],
            label = classes[i],
            left=x_sum
        )
    x_sum += class_data[i]
    
ax.set_xlabel('No. samples')
ax.set_title('Class distribution in Clients; Î±={}'.format(alpha))
ax.autoscale(enable=True, axis='x', tight=True)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
fig.savefig("images/mnist_lda_partitioning_alpha_1.png", bbox_inches='tight', pad_inches=0)
```
